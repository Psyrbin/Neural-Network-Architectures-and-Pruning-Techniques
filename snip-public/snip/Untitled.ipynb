```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle
```

```python
dicts = []
for seed in range(1000):
    with open('masks/sc_sparsity=0.9_seed=' + str(seed), 'rb') as f:
        dicts.append(pickle.load(f))
len(dicts)
```

```python
w1s = []
for d in dicts:
    w1s.append(d['w1'])
```
